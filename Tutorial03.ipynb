```python
 import numpy as nump
```

1. Create a numpy array of size 2 x 2

```python
array = nump.array([[1,2],[4,5]])
print (array)
```

2. Create a random numpy array of size 3 x 4

```python
array2 = nump.random.rand(3,4)
print (array2)
```

3. Create a numpy array of ones and a numpy array of zeroes

```python
array_ones = nump.ones((2,2))
array_zeros = nump.zeros((2,2))

print ("Array of ones : \n ", array_ones)
print ("Array of zeros : \n ", array_zeros)
```

4. Create a random numpy array of size 6x3 and transpose it

```python
random_array = nump.random.randint(1,100,(6,3))
print ("Original array 6x3 : \n", random_array)
print ("\n")
print ("Transpose array 6x3 : \n", random_array.T)
```

5. Create two random numpy array of size n x p and p x m and then compute the dot product

```python
# n=2, p=3, m=4

M = nump.random.randint(1,50,(2,3))
Z = nump.random.randint(1,50,(3,4))

print ("Array n x p : \n", M)
print ("\n")
print ("Array m x p : \n", Z)

dot_product = nump.dot(M,Z)

print ("\n'")
print ("Dot product : \n", dot_product)
```
